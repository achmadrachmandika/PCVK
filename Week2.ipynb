##2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar
img2 = cv.imread('/content/drive/MyDrive/PCVK/Week2/kitten.jpg')

# Mengubah menjadi warna re-blue (merah dan biru)
img_rb = img2.copy()
img_rb[:, :, 1] = 0  # Mengatur komponen hijau menjadi 0

# Mengubah menjadi warna green-blue (hijau dan biru)
img_gb = img2.copy()
img_gb[:, :, 2] = 0  # Mengatur komponen merah menjadi 0

# Menampilkan kedua gambar secara bersamaan menggunakan subplots
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.title('Gambar Asli')


plt.subplot(132)
plt.imshow(cv.cvtColor(img_rb, cv.COLOR_BGR2RGB))
plt.title('Warna Re-Blue')


plt.subplot(133)
plt.imshow(cv.cvtColor(img_gb, cv.COLOR_BGR2RGB))
plt.title('Warna Green-Blue')


plt.show()
```


##5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
# Buat gambar kosong (hitam) sebagai dasar
width, height = 1000, 800
black_img2 = np.zeros((height, width, 3), dtype=np.uint8)

# Tentukan koordinat dan properti kotak
boxes = [
    ((50, 100), (200, 200), (0, 0, 255)),  # Kotak 1 (Merah)
    ((300, 300), (400, 400), (0, 255, 0)),  # Kotak 2 (Hijau)
    ((500, 100), (700, 200), (255, 0, 0)),  # Kotak 3 (Biru)
    ((100, 400), (250, 500), (255, 255, 0)),  # Kotak 4 (Kuning)
    ((500, 400), (700, 600), (255, 0, 255))  # Kotak 5 (Ungu)
]

# Gambar kotak-kotak pada gambar
for box in boxes:
    pt1, pt2, color = box
    cv.rectangle(black_img2, pt1, pt2, color, thickness=5)

# Tampilkan gambar yang telah dimodifikasi
plt.imshow(cv.cvtColor(black_img2, cv.COLOR_BGR2RGB))
plt.show()
```
